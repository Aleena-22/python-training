```python
# Write a decorator called @log_function that prints the name of the 
# function being called and its arguments before executing it.

# @log_function
# def greet(name):
#     print(f"Hello, {name}!")

# greet("Alice")

# Output: 
# Calling function: greet 
# Arguments: 'Alice'
# Hello, Alice. How are you today?

def log_function(func):
    def wrapper(*args):
        print(f"Calling function: {func.__name__}")
        print(f"Arguments: {', '.join(map(repr, args))}")
        return func(*args)
    return wrapper

@log_function
def greet(name):
    print(f"Hello, {name}! How are you today?")

greet("Alice") 
```

```python
# Write a decorator @double_result that doubles the result returned by any function.

# Example:

# @double_result
# def add(x, y):
#     return x + y

# print(add(3, 4))  # Output: 14

def double_result(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs) * 2
    return wrapper

@double_result
def add(x, y):
    return x + y

print(add(3, 4))
```

```python
# Write a decorator @timer that prints how long a function takes to execute.

# import time

# @timer
# def slow_function():
#     time.sleep(1)
#     print("Done!")

# slow_function() # 1 seconds

import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"Time taken: {time.time() - start} sec")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(1)
    print("Done!")

slow_function()
```
